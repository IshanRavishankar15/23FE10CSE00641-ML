##Descriptive Statistical Analysis

Aim: Apply descriptive statistical techniques to summarize and interpret
datasets.

```python
# Generate descriptive statistics for all numeric features
df.describe()
```

```python
# Summarize statistics grouped by health status
df.groupby("status").describe()
```

```python
# Compute mean of each feature
df.mean(numeric_only=True)
```

```python
# Compute median of each feature
df.median(numeric_only=True)
```

```python
# Measure variability using standard deviation
df.std(numeric_only=True)
```

```python
# Identify range of feature values
df.min(numeric_only=True), df.max(numeric_only=True)
```

```python
# Compute variance of features
df.var(numeric_only=True)
```

```python
# Measure skewness of feature distributions
df.skew(numeric_only=True)
```

```python
# Measure kurtosis of feature distributions
df.kurtosis(numeric_only=True)
```

```python
# Compute quartiles for numeric features
df.quantile([0.25, 0.5, 0.75], numeric_only=True)
```

```python
# Compute correlation between features
df.corr(numeric_only=True)
```

```python
# Visualize mean differences between classes
df.groupby("status").mean(numeric_only=True).T.plot(kind="bar", figsize=(16,6))
plt.ylabel("Mean Value")
plt.title("Mean Feature Comparison by Status")
plt.show()
```

```python
# Visualize spread of key features
df[["Fo", "HNR", "PPE"]].plot(kind="box", figsize=(8,5))
plt.title("Feature Spread")
plt.show()
```

```python
# Visualize distributions of selected features
df[["Fo", "HNR", "PPE"]].hist(bins=20, figsize=(10,6))
plt.tight_layout()
plt.show()
```
