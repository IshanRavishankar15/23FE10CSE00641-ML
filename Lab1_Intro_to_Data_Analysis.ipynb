#Lab 1 - Introduction to Python for Data Analysis

### Parkinson's Disease Dataset –
Basic exploration of the Parkinson's voice measurement dataset.


```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# Load dataset
df = pd.read_csv("parkinsons.csv")

df.head()
```

```python
# Basic dataset info
print("Shape:", df.shape)
df.info()
```

```python
# View column names
df.columns
```

```python
# Numerical summary
df.describe()
```

```python
# Target distribution
sns.countplot(x="status", data=df)
plt.title("Parkinson's Status Distribution")
plt.show()
```

```python
# Correlation heatmap (numeric features only)
plt.figure(figsize=(14, 10))
sns.heatmap(df.select_dtypes(include=np.number).corr(),
            cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Example feature comparison
sns.boxplot(x="status", y="HNR", data=df)
plt.title("HNR vs Status")
plt.show()
```

```python
# Simple pandas plot for a few numeric features
df.select_dtypes(include=np.number)[
    ["MDVP:Fo(Hz)", "MDVP:Fhi(Hz)", "MDVP:Flo(Hz)"]
].plot(figsize=(10, 5))

plt.title("Basic Pandas Line Plot of Selected Features")
plt.xlabel("Sample Index")
plt.ylabel("Value")
plt.show()
```

```python
# Scatter plot colored by Parkinson's status
df.plot.scatter(
    x="MDVP:Fo(Hz)",
    y="HNR",
    c="status",
    cmap="coolwarm",
    figsize=(7, 5)
)

plt.title("MDVP:Fo(Hz) vs HNR by Status")
plt.show()
```
